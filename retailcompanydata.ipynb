There are a total of 37 null values in the Review Rating column, so we need to decide how to handle them.

We could use either the mean or the median for imputation. However, I prefer using the median because **the mean is sensitive to outliers, which can skew the distribution of the data**.

**The median is more robust and preserves the central tendency of the ratings without being influenced by extreme values**.

we will find out median review rating within each category and we will place it with the missing value of that category

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

If we observe the column headings, they contain a mix of **uppercase **and **lowercase** letters along with spaces. This can lead to problems while writing queries in **MySQL** and when accessing columns in **Python**, as such names require extra handling **(quotes, brackets, or special syntax).**

To avoid these issues and follow best practices, we standardize all column names by converting them to **lowercase **and replacing spaces with **underscores**.

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```
